```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df= pd.read_csv("student_performance.csv")
df.head()
```

```python
df.describe()
```

```python
sns.pairplot(df, hue="Marks_Obtained", palette="bright", markers="o")
plt.scatter(df["Marks_Obtained"],df["Hours_Studied"], color="k")
plt.show()
```

```python
train,test=train_test_split(df, test_size=0.2, random_state=42)
```

```python
model=LinearRegression()
x_train=train[["Marks_Obtained"]]
y_train=train[["Hours_Studied"]]
model.fit(x_train,y_train)
```

```python
x_test=test[["Marks_Obtained"]]
y_test=test[["Hours_Studied"]]
y_pre=model.predict(x_test)
```

```python
plt.scatter(x_test,y_test, marker="o")
plt.scatter(y_test,y_pre, marker="*")
plt.show()
```

```python
# mse=mean_squared_error(x_train,y
mae=mean_absolute_error(y_test,y_pre)
mse=mean_squared_error(y_test,y_pre)
r2=r2_score(y_test,y_pre)
print(r2)
print(mae)
print(mse)
```
