```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
# import kagglehub
```

```python
df= load_iris()
data= pd.DataFrame(data=df.data, columns=df.feature_names)
data.head()
# print(df.head())
```

```python
# data= sns.load_dataset("anagrams")
print(data.head())
print(data.info())
print(data.describe())
```

```python
sns.pairplot(data, palette="bright", hue='petal length (cm)')
plt.show()
```

```python
sns.jointplot(data=data, x="petal width (cm)", y= "sepal length (cm)")
plt.show()
```

```python
train,test= train_test_split(data, test_size=0.2, random_state=42)
```

```python
model= LinearRegression()
x_train=train[["sepal length (cm)","petal width (cm)"]]
y_train=train[['petal length (cm)']]
model.fit(x_train,y_train)
```

```python
x_test=train[["sepal length (cm)" , "petal width (cm)"]]
y_test=train[['petal length (cm)']]
y_pred=model.predict(x_test)
# print("Predicted value: ", pred_y)
```

```python
plt.scatter(x_test["sepal length (cm)"], y_test, color='blue', label='Actual')
plt.scatter(x_test["petal width (cm)"], y_test, color='k', label='Actual')
plt.scatter(x_test["sepal length (cm)"], y_pred, color='red', label='Predicted')
plt.scatter(x_test["petal width (cm)"], y_pred, color='y', label="Predicted")
plt.show()
```

```python
mae=mean_absolute_error(y_test, y_pred)
mse=mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(r2)
print(mae)
print(mse)
```

```python
from sklearn.preprocessing import StandardScaler

# 1. Select the input features you plan to scale
X = data[["sepal length (cm)", "petal width (cm)"]]
y = data[["petal length (cm)"]]

# 2. Scale the features using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Proceed with train-test split on scaled data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 4. Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# 5. Predict and evaluate
y_pred = model.predict(X_test)
r2= r2_score(y_test,y_pred)
```
