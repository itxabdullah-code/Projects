```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

pd.lineplot is use to create a graph from x-aixs and y

```python
velocity=[1,3,5,6,7,8,12,24]
accelration=[2,4,6,8,10,12,14,16]
df=pd.DataFrame({"V":velocity, "A":accelration})
print(df)
sns.lineplot(x="V",y="A",data=df)
```

```python
velocity=[1,3,5,6,7,8,12,24]
accelration=[2,4,6,8,10,12,14,16]
data1={"v":velocity,"a":accelration}
# data= pd.DataFrame(velocity)
# # data.add(accelration)
sns.lineplot(x="v" , y="a", data=data1, weights=200)
```

```python

# Example data for different states
x = [1, 2, 3, 4, 5]
state1 = [10, 15, 13, 17, 20]
state2 = [8, 12, 11, 14, 18]

# Plotting lines with labels
plt.plot(x, state1, label='California', color="k")
plt.plot(x, state2, label='Texas', color="g")

# Add title and labels
plt.title("State Data Comparison")
plt.xlabel("Time")
plt.ylabel("Values")

# Show the legend
plt.legend("VT")

# Display the plot
plt.show()
```

```python
print(sns.get_dataset_names())
iris = sns.load_dataset("iris")
print(iris)
```

```python
sns.histplot(data=iris, x="sepal_length", kde=True)
plt.title("Distribution of Sepal Lengths")
plt.xlabel("Sepal Lenght")
plt.ylabel("Sepal Height")
# plt.grid()
plt.show()
```

```python
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue="species")
plt.title("Sepal vs Petal Length by Species")
plt.show()
```

```python
sns.boxplot(data=iris, x="species", y="petal_width")
plt.title("Petal Width Across Species")
plt.show()
```

```python
sns.set()
sns.boxplot(data=iris, x="petal_length", y="petal_width", label= "petal")
```

```python
sns.violinplot(data=iris, x="species", y="sepal_width")
plt.title("Sepal Width Distribution by Species")
plt.show()
```

```python
sns.scatterplot(data=iris, x="petal_length", y="petal_width", hue="species")
plt.title("Simplest Iris Insight")
plt.show()
```

```python
# Style settings
sns.set_style("whitegrid")
sns.set_palette("husl")  # Colorful palette
plt.figure(figsize=(10,6))  # Larger plots
sns.scatterplot(data=iris, x="petal_length", y="petal_width", hue="species")
plt.title("Simplest Iris Insight")
```

```python
corr = iris.drop(columns="species").corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```
